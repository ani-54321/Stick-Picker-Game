```python
f = open('gameData.txt','r')
name = []
wins = []
for i in f:
    a = i.find(':')
    b = i.find(' ', a)
    name.append(i[:a])
    wins.append(int(i[a+1:b]))
    
#name
```

```python
file = open('gameDateData.txt','r')
date = []
time = []
for l in file:
    st = l.find(' ')
    nd = l.find(' ', st+1)
    date.append(l[:st])
    time.append(l[st+1:nd])

new_date = []
new_time = []
j = 0
for i in range(len(wins)):
    if i%4>=0 and i%4!=3:
        new_date.append(date[j])
        new_time.append(time[j])
    elif i%4==3:
        new_date.append(date[j])
        new_time.append(time[j])
        j+=1
        

#new_time
```

```python
counts = []
for n in new_names:
    counts.append(name.count(n))
    
#counts
```

```python
max = np.array(new_wins)
max = max.max()

min = np.array(new_wins)
min = min.min()
print('Minimum wins are :', min, ' Maximum wins are :', max)
```

```python
levels = []
for j in range(len(new_wins)):
    levels.append(new_wins[j]+counts[j])
    
levels
```

```python
loses = []
for i in range(len(new_wins)):
    loses.append(levels[i]-new_wins[i])

loses
```

```python
percent_win = []
for i in range(len(new_wins)):
    percent_win.append(str(round(new_wins[i]*100/levels[i],2))+'%')
    
#percent_win
```

```python
percent_lose = []
for i in range(len(loses)):
    percent_lose.append(str(round(loses[i]*100/levels[i],2))+'%')

#percent_lose
```

```python
suggestions = []
for i in range(len(levels)):
    if 1<=levels[i]<3:
        suggestions.insert(i, f'Play some more matches {new_names[i]}')
    elif levels[i]>3 and new_wins[i]<3:
        suggestions.insert(i, f'Play with concentration {new_names[i]}')
    elif 1<levels[i]<4 and new_wins[i]==0:
        suggestions.insert(i, f'Dedicate yourself {new_names[i]}')
    elif levels[i]>4 and new_wins[i]>levels[i]-3:
        suggestions.insert(i, f'Nice work {new_names[i]}')
    else:
        suggestions.insert(i, f'Keep it up {new_names[i]}')
        
#suggestions
```

```python
df = pd.DataFrame([new_names,new_wins,loses,levels,percent_win,percent_lose,counts,suggestions],['Players','Total Wins','Total Losses','Matches Played','Percent Wins','Percent Losses','Enrolled','Suggestions'],range(1,len(new_wins)+1))
df = df.T
df
```
